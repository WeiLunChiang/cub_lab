## 消費資料產生器demo

```python
import os
from langchain_core.runnables import RunnablePassthrough, RunnableParallel
from langchain_openai import AzureChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import JsonOutputParser
from langchain_core.runnables import RunnableLambda
from utils_vectordb import get_or_create_http_chromadb
from utils_retriever import get_metadata_runnable, get_sql_querys, RetrieveWithScore
from module_1_sql_query_extract import template_2
from chat_momery import runnable_with_history
from icecream import ic as print
from create_and_insert_sqlite import query_data_from_rdb
```

### 0. 定義參數

```python
user_input = "過去一年我在蝦皮花了多少錢?"
today = "2024-05-01"
user_id = "A"
k=3
score_threshold=0
collection_name = "collect_cubelab_qa_lite"
rdb_path = "cubelab_txn.db"
```

### 1. 定義元件

```python
model = AzureChatOpenAI(
    openai_api_version=os.environ["AZURE_OPENAI_API_VERSION"],
    azure_deployment=os.environ["AZURE_OPENAI_CHAT_DEPLOYMENT_NAME"],
)

vectorstore = get_or_create_http_chromadb(collection_name=collection_name)

retriever = RetrieveWithScore(vectorstore, k=k, score_threshold=score_threshold)

prompt = ChatPromptTemplate.from_template(template_2)
```

### 2. 定義chain

```python
#挖空用戶問題, 拿到挖空後用戶問句及參數
json_parser_chain = prompt | model | JsonOutputParser()

#query VDB拿到前K接近的資料及對應metadata
retriever_chain = (
    RunnableLambda(lambda response: response["modify_query"]) #挖空後的用戶問句
    | retriever
    | {  
        "SQL": get_metadata_runnable("SQL1", "SQL2", "SQL3"), #對應sql查詢語句(挖空)
        "標準問題": get_metadata_runnable("問題類別"), #標準問題
        "category": get_metadata_runnable("category"), #問題類別A、B、C、D
        "score": get_metadata_runnable("score"), #相似度(最小值)
    }
)


chain_1 = (
    {
        "question": RunnablePassthrough(),
        "today": RunnableLambda(lambda x: today),
    }
    | json_parser_chain
    | {
        "keys": RunnablePassthrough(),  # 解析出來的參數
        "retriever": retriever_chain,
    }
)
```

### 3.1 解析用戶問題

```python
user_input = "過去一年我在蝦皮花了多少錢?"
response = chain_1.invoke(user_input)
response
```
