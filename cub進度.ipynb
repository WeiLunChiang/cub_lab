```python
import pandas as pd
```

```python

df = pd.read_csv(
    "json_files_from_query_ner/similarity_search_result/merged_data_with_extra_q.csv"
)
df["排序"] = df.groupby(["問題類別", "原始問題"])["相似度"].rank(
    ascending=False, method="dense"
)


df_val = pd.read_csv("qa_set_with_sql_lite.csv")[["變形問題", "category"]]
df_val["變形問題"] = df_val["變形問題"].apply(lambda x: x.replace("\xa0", "").lstrip())
df_val.rename({"變形問題": "document"}, axis=1, inplace=True)

df_ner = df[df["排序"] == 1].merge(df_val, on="document", how="left")
df_ner.groupby('問題類別',group_keys=False).apply(lambda x: (x.問題類別 == x.category).mean())
```
